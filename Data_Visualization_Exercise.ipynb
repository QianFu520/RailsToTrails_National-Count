```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_excel("/content/RailsToTrails_National Count Data_week.xlsx")
df.head()
```

```python
df.dtypes
```

```python
Pedestranis = df["Pedestrians (14 counters)"].values
Bikes = df["Bikes (14 counters)"].values
Week = df["Week"].values
```

```python
plt.style.use("seaborn")
fig, axes = plt.subplots(nrows = 1, ncols = 1, figsize = (15, 10))
axes.plot(Week, Bikes, c = "green", marker = ".", markersize = 10, label = "Bikes")
axes.plot(Week, Pedestranis, c = "magenta", marker = ".", markersize = 10, label = "Pedestranis")
axes.legend(fontsize = 16)
axes.grid(axis= "y", c = "k", alpha = .9, linestyle = "-")
axes.set_ylim(bottom=0,top=180000)
plt.xticks(fontsize = 16)
plt.yticks(fontsize = 16)
axes.set_xlabel("Weeks", fontsize = 20)
axes.set_ylabel("Total of Miles", fontsize = 20)
axes.set_title("Total of Miles Over the Weeks", fontsize = 20)
```
